### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def get_probs(Q_s, nA, epsilon):
    """ Get Probability of each action based on current Q table"""
    policy_s = np.ones(nA)*(epsilon/nA)
    action = np.argmax(Q_s)
    policy_s[action] = 1 -  epsilon + (epsilon/nA)
    return policy_s
````

````python
def sarsa_policy(env, Q, epsilon, nA, alpha, gamma):
    """ Get the SAR pair from sarsa policy"""
    episode = []
    state = env.reset()
    while True:
        action = np.random.choice(np.arange(nA), p = get_probs(Q[state],nA, epsilon)) if state in Q else env.action_space.sample()
        next_state, reward, done, info = env.step(action)
        #The above episode completes the SARSA step, now since we have sarsa
        #we can choose to take the action given by policy determined by Q
        next_action = np.random.choice(np.arange(nA), p = get_probs(Q[next_state],\
                                                                  nA, epsilon)) if next_state in Q else env.action_space.sample()
        
        current_q_value = Q[state][action] if state in Q else 0
        
        next_q_value = Q[next_state][next_action] if next_state in Q else 0
        
        Q[state][action] = current_q_value + alpha*(reward + \
                                                    gamma*next_q_value - current_q_value)
        
        episode.append((state, action, reward))
        state = next_state         
        if done:
            break
    return episode, Q

````

````python
def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        
        _episodes, Q = sarsa_policy(env, Q, 1/i_episode, env.nA, alpha, gamma)

        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        ## TODO: complete the function
        
    return Q
````

### Part 3: TD Control: Expected Sarsa

(_Feel free to define additional functions to help you to organize your code._)

````python
def sarsa_policy(env, Q, epsilon, nA, alpha, gamma):
    """ Get the SAR pair from sarsa policy"""
    episode = []
    state = env.reset()
    while True:
        action = np.random.choice(np.arange(nA), p = get_probs(Q[state],nA, epsilon)) if state in Q else env.action_space.sample()
        next_state, reward, done, info = env.step(action)
        #The above episode completes the SARSA step, now since we have sarsa
        #we can choose to take the action given by policy determined by Q
        
        current_q_value = Q[state][action] if state in Q else 0
        
        policy_next_state = get_probs(Q[next_state],nA, epsilon) if next_state in Q else np.ones(nA)*(1/nA)
        
        next_q_value = np.dot(policy_next_state, Q[next_state]) if next_state in Q else 0
                
        Q[state][action] = current_q_value + alpha*(reward + \
                                                    gamma*next_q_value - current_q_value)
        
        episode.append((state, action, reward))
        state = next_state         
        if done:
            break
    return episode, Q
````
